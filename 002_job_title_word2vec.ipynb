```python
# coding: utf-8
import multiprocessing
import gensim
import sys
import re

def my_filter(x): # 是否含有特殊字符
	match1 = re.fullmatch(r"^[^0-9a-zA-Z\u4e00-\u9f5a]+", x)
	match2 = re.fullmatch(r"^[0-9]+", x)
	match3 = x == ""
	if (match1 or match2 or match3):
		return False
	else:
		return True
        
def reconstruct_all_title_list(filePath):
	all_title_list = []
	with open(filePath,"rb") as fin:
		line = fin.readline().decode('utf-8').strip()
		while(line):
			title_list = line.split(sep=" ")
			title_list = list(filter(my_filter, title_list))
			all_title_list.append(title_list)
			line = fin.readline().decode('utf-8').strip()
	return all_title_list


def train_title_word2vec(in_file, out_file):
	all_title_list = reconstruct_all_title_list(in_file)
	# 使用word2vec训练
	cores = multiprocessing.cpu_count()
	print("using all " + str(cores) + " cores")
	print("Training word2vec model...")
	model = gensim.models.Word2Vec(sentences=all_title_list, size=150, min_count=10, window=2, workers=cores)
	print("Saving model...")
	model.save(out_file)
```
