```python
# coding:utf-8
import gensim
model_path = "./model/job_title_word2vec.model"
model = gensim.models.Word2Vec.load(model_path)
```

```python
def show_word_most_similar(word, model, topn):
    result_word2vec_list = model.most_similar(word, topn=topn)
    i = 0
    for wv in result_word2vec_list:
        i+=1
        print("%-5s%-10s%-10s" % (i, wv[0], wv[1]))
```

```python
show_word_most_similar("java", model, 10)
```

```python
index2word = {}
positions = [line.encode('utf-8').decode('utf-8-sig').strip() for line in open("./positions/positions0928.txt", 'r', encoding='utf-8')]
for i in range(len(positions)):
    index2word[i] = positions[i]
```

```python
len(index2word)
```

```python
positions_w2v_list = []
for position in positions:
    positions_w2v_list.append(wv[position].tolist())
```

```python
len(positions_w2v_list)
```

```python
import numpy as np
X = np.array(positions_w2v_list)
```

```python
X.shape
```

```python
# kmeans聚类
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
for i in range(1000):
    kmeans = KMeans(n_clusters=110, init='k-means++', n_init=10).fit(X)
    score = silhouette_score(X, kmeans.labels_)
    print("%-5s%-10s" % (i, score))
    if(score > 0.2): 
        break
```

```python
# 寻找最佳ncluster
for ncluster in range(100, 150):
    kmeans = KMeans(n_clusters=ncluster, init='k-means++', n_init=5).fit(X)
    score = silhouette_score(X, kmeans.labels_)
    print("%-5s%-10s" % (ncluster, score))
```

```python
# 将聚类结果保存下来
n_clusters = 115
fouts = []
for i in range(n_clusters):
    fout= open("./positions_kmeans_result/0928/topic"+str(i+1)+".txt", "w", encoding="utf-8")
    fouts.append(fout)
    
index = 0
for label in kmeans.labels_:
    fouts[label-1].write(index2word[index])
    fouts[label-1].write("\n")
    index += 1
for fout in fouts:
    fout.close()
```

```python
# 将聚类结果保存为一个文件
with open("./positions_kmeans_result/positions_labeled_115_0928.txt", "w", encoding="utf-8") as fo:
    index = 0
    for label in kmeans.labels_:
        # print(index2word[index],label)
        fo.write(index2word[index] + " " + str(label))
        fo.write("\n")
        index += 1
```
